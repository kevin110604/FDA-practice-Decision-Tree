### Create data

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO   
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score
import pydotplus
```

```python
sex_choice = ['male', 'female']
maj_choice = ['science', 'art']
yes_no = [True, False]

age = np.random.choice(100, size=1000, replace=True)
sex = np.random.choice(sex_choice, size=1000, replace=True)
stu = np.random.choice(yes_no, size=1000, replace=True)
maj = np.random.choice(maj_choice, size=1000, replace=True)
sal = np.random.choice(100000, size=1000, replace=True)
rcb = np.random.choice(yes_no, size=1000, replace=True)
rdb = np.random.choice(yes_no, size=1000, replace=True)
```

```python
df = pd.DataFrame(data={'Age': age, 'Sex': sex, 'Student': stu, 
                        'Major': maj, 'Salary': sal, 'Raised_a_cat_before': rcb, 
                        'Raised_a_dog_before': rdb, 'Preference': ''})
df.head()
```

```python
def f(x):
    if x[0] < 30:
        if x[2] == True:
            if x[5] == True:
                return 'cat'
            else:
                if x[1] == 'male':
                    if x[3] == 'science':
                        return 'dog'
                    else:
                        return 'cat'
                else:
                    return 'cat'
        else:
            if x[5] == True:
                return 'cat'
            else:
                if x[1] == 'male':
                    if x[4] >= 30000:
                        if x[6] == True:
                            return 'dog'
                        else:
                            return 'cat'
                    else:
                        return 'dog'
                else:
                    if x[4] >= 45000:
                        if x[6] == True:
                            return 'dog'
                        else:
                            return 'cat'
                    else:
                        return 'dog'
    else:
        if x[0] < 60:
            if x[1] == 'male':
                if x[4] >= 80000:
                    if x[5] == True:
                        return 'cat'
                    else:
                        return 'dog'
                else:
                    return 'dog'
            else:
                if x[3] == 'science':
                    if x[5] == True:
                        return 'cat'
                    else:
                        return 'dog'
                else:
                    return 'cat'
        else:
            if x[1] == 'male':
                return 'dog'
            else:
                if x[5] == True:
                    return 'cat'
                else:
                    return 'dog'

#                       0      1      2          3        4
df['Preference'] = df[['Age', 'Sex', 'Student', 'Major', 'Salary',
                       'Raised_a_cat_before', 'Raised_a_dog_before']].apply(f, axis=1)
#                       5                      6
df
```

## Step 2

Use the data generated in step 1 to build a decision tree

### Data preprocessing

```python
sex_map = {'male': 0, 'female': 1}
df['Sex'] = df['Sex'].map(sex_map)
df.head()
```

```python
maj_map = {'art': 0, 'science': 1}
df['Major'] = df['Major'].map(maj_map)
df.head()
```

```python
true_false_map = {True: 0, False: 1}
df['Student'] = df['Student'].map(true_false_map)
df.head()
```

```python
df['Raised_a_cat_before'] = df['Raised_a_cat_before'].map(true_false_map)
df.head()
```

```python
df['Raised_a_dog_before'] = df['Raised_a_dog_before'].map(true_false_map)
df.head()
```

### Seperate data into training set and testing(validation) set

```python
df_train = df[:800]
df_train.head()
```

```python
df_test = df[800:]
df_test.head()
```

### Seperate attributes and answer

```python
y = df_train['Preference'].values
df_train = df_train.drop('Preference', 1)
y
```

### Build Decision Tree

```python
dtree=DecisionTreeClassifier(max_depth=6)
dtree.fit(df_train, y)

dot_data = StringIO()
export_graphviz(dtree, 
                out_file=dot_data,  
                filled=True, 
                feature_names=list(df_train),
                class_names=['cat', 'dog'],
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_pdf("cat_dog_tree.pdf")
```

### My cat_dog_tree

### Calculate accuracy

```python
y_test = df_test['Preference'].values
X_test = df_test.drop('Preference', 1)
y_predict = dtree.predict(X_test)
y_predict
```

```python
accuracy_score(y_test, y_predict)
```

### 相似程度

大致上來說，features 的 thresholds 都蠻像的，不一樣的地方在於上下層的位置。很多時候，是否有養過貓的這個 feature 甚至會跑到最頂端，這可能跟我在設計 rules 的時候，把每個有養過貓的人都歸類為貓派有關，所以 `DecisionTreeClassifier` 才會認為這個 feature 相當重要。那為什麼是否有養過狗這個 feature 沒有同樣的情形發生? 因為在原本的 rules 當中，是否有養過狗這個 feature 都是最後才考慮的因素。

除了上述因素，造成 decision tree 與 rules 不一樣的原因可能還有我設變數 max_depth=6 的關係。rules 裡頭其實只有少數幾個分支有到 6 層，造成 decision tree 有 overfitting 的情形，多考慮了 rules 裡沒有的條件。
