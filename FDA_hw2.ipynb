```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO   
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score
import pydotplus
```

## Step 1

Design features/attributes for your problem and a set of rules to classify data

Question: What kinds of people would like cats more than dogs?

Features: Age, Sex, Student, Major, Raised_a_cat_before, Raised_a_dog_before, Salary

```python
sex_choice = ['male', 'female']
maj_choice = ['science', 'art']
yes_no = [True, False]

age = np.random.choice(100, size=1000, replace=True)
sex = np.random.choice(sex_choice, size=1000, replace=True)
stu = np.random.choice(yes_no, size=1000, replace=True)
maj = np.random.choice(maj_choice, size=1000, replace=True)
sal = np.random.choice(100000, size=1000, replace=True)
rcb = np.random.choice(yes_no, size=1000, replace=True)
rdb = np.random.choice(yes_no, size=1000, replace=True)
```

```python
df = pd.DataFrame(data={'Age': age, 'Sex': sex, 'Student': stu, 
                        'Major': maj, 'Salary': sal, 'Raised_a_cat_before': rcb, 
                        'Raised_a_dog_before': rdb, 'Preference': ''})
df.head()
```

```python
def f(x):
    if x[0] < 30:
        if x[2] == True:
            if x[5] == True:
                return 'cat'
            else:
                if x[1] == 'male':
                    if x[3] == 'science':
                        return 'dog'
                    else:
                        return 'cat'
                else:
                    return 'cat'
        else:
            if x[5] == True:
                return 'cat'
            else:
                if x[1] == 'male':
                    if x[4] >= 30000:
                        if x[6] == True:
                            return 'dog'
                        else:
                            return 'cat'
                    else:
                        return 'dog'
                else:
                    if x[4] >= 45000:
                        if x[6] == True:
                            return 'dog'
                        else:
                            return 'cat'
                    else:
                        return 'dog'
    else:
        if x[0] < 60:
            if x[1] == 'male':
                if x[4] >= 60000:
                    return 'cat'
                else:
                    return 'dog'
            else:
                if x[3] == 'science':
                    if x[5] == True:
                        return 'cat'
                    else:
                        return 'dog'
                else:
                    return 'cat'
        else:
            return 'dog'

#                       0      1      2          3        4         5                      6
df['Preference'] = df[['Age', 'Sex', 'Student', 'Major', 'Salary', 'Raised_a_cat_before', 'Raised_a_dog_before']].apply(f, axis=1)
df
```

## Step 2

Use the data generated in step 1 to build a decision tree

```python
sex_map = {'male' : 0, 'female' :1}
df['Sex'] = df['Sex'].map(sex_map)
df.head()
```

```python
maj_map = {'science' : 0, 'art' :1}
df['Major'] = df['Major'].map(maj_map)
df.head()
```

```python
df_train = df[:800]
df_train.head()
```

```python
df_test = df[800:]
df_test.head()
```

```python
y = df_train['Preference'].values
df_train = df_train.drop('Preference', 1)

y
```

```python
dtree=DecisionTreeClassifier(max_depth=6)
dtree.fit(df_train, y)

dot_data = StringIO()
export_graphviz(dtree, 
                out_file=dot_data,  
                filled=True, 
                feature_names=list(df_train),
                class_names=['cat', 'dog'],
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_pdf("cat_dog_tree.pdf")
```

```python
y_test = df_test['Preference'].values
X_test = df_test.drop('Preference', 1)

y_predict = dtree.predict(X_test)

y_predict
```

```python
accuracy_score(y_test, y_predict)
```
